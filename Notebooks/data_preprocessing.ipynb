```python
import pandas as pd
import json
```

```python
df=pd.read_csv("../Resources/austin_housing.csv")
df
```

```python
df.columns
```

```python
#numOfPrimarySchools', 'numOfElementarySchools', 'numOfMiddleSchools', 'numOfHighSchools' sum these into "numOfSchools"
df["numOfSchools"]=df['numOfPrimarySchools']+df['numOfElementarySchools']+df['numOfMiddleSchools']+df['numOfHighSchools']
df[["numOfSchools"]]
```

```python
#remove homeImage, avgSchoolSize, MedianStudentsPerTeacher, latest_salemonth, latest_saleyear, zpid, numOfPhotos
df.drop(columns=[ "homeImage", 
                 "avgSchoolSize", 
                 "MedianStudentsPerTeacher", 
                 "latest_salemonth", 
                 "latest_saleyear",
                 "latest_saledate",
                 "zpid",
                 "avgSchoolDistance",
                 "numOfPhotos",
                 "hasAssociation",
                 'numOfPrimarySchools',
                 "latestPriceSource",
                 'numOfElementarySchools', 
                 'numOfMiddleSchools',
                 'parkingSpaces',
                 'numOfHighSchools'], inplace=True)

df.head()
```

### Filter out 0 bedrooms, too many bath and/or beds, too expensive/cheap

```python
prices=df.sort_values(by="latestPrice")

prices[["latestPrice"]].head(30)
```

```python
df=df.loc[(df["latestPrice"]>45000) & 
          (df["latestPrice"]<6500000) &
          (df["numOfBedrooms"]>0) &
          (df["numOfBathrooms"]>0) &
          (df["numOfBathrooms"]!= 27) &
          (df["numOfBathrooms"]!= 1.75) &
          (df["numOfBathrooms"]!= 1.70) &
          (df["numOfBedrooms"]!= 20) &
          (df["numOfBathrooms"]!= 2.75)]
df.reset_index(drop=True,inplace=True)
```

```python
df.columns
```

```python
df.to_csv("../Resources/austin_housing_reduced.csv",index=False)
```

## Create GEOJSON file

```python
geo={"type": "FeatureCollection","features": []}

for i in df.index:
    one_feat={"type": "Feature",
     "geometry": {
        "type": "Point",
        "coordinates": [df["longitude"].iloc[i],df["latitude"].iloc[i]]
      },
     "properties": {"Address":df["streetAddress"].iloc[i],
                  "zipcode":df["zipcode"].iloc[i],
                    "garageSpaces":df["garageSpaces"].iloc[i],
                       "numOfStories":df["numOfStories"].iloc[i],
                                            "homeType":df["homeType"].iloc[i],
                                                                 "numOfBathrooms":df["numOfBathrooms"].iloc[i],
                                                                                        "numOfBedrooms":df["numOfBedrooms"].iloc[i]
                                                    
                   }
    }
    geo["features"].append(one_feat)
```

```python
geo
```

```python
with open("../Resources/austin.geojson", "w") as outfile:
    json.dump(geo, outfile)
```
